```python
import numpy as np
```

```python
arr = np.array([3.7, -1.2, -2.6, 0.5, 12.9, 10.1])
print("ndim:", arr.ndim)
print("shape:", arr.shape)
print("dtype:", arr.dtype)
arr
```

```python
arr.astype(np.int32)
```

```python
arr > 0
```

```python
arr[arr > 0]
```

```python
np.where(arr > 0, 2, -2)
```

## random

```python
np.random.seed(42)
rng = np.random.default_rng(seed=42)
type(rng)
```

```python
rng.integers(low=0, high=10, size=(3, 4))
```

```python
rng.standard_normal(size=(3, 4))
```
